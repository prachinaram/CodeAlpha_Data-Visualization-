```python
plt.figure(figsize=(8,5))
sns.histplot(data["Daily_Screen_Time(hrs)"], kde=True)
plt.title("Distribution of Daily Screen Time")
plt.xlabel("Screen Time (hrs)")
plt.ylabel("Count")
plt.show()
```

```python
#YouTube and TikTok are the most commonly used platforms.

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

plt.figure(figsize=(8,5))
sns.countplot(x="Social_Media_Platform", data=data)
plt.title("Most Used Social Media Platforms")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x="Gender", y="Stress_Level(1-10)", data=data)
plt.title("Average Stress Level by Gender")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="Daily_Screen_Time(hrs)", y="Stress_Level(1-10)", data=data)
plt.title("Screen Time vs Stress Level")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(
    x="Sleep_Quality(1-10)",
    y="Happiness_Index(1-10)",
    data=data
)
plt.title("Average Happiness by Sleep Quality")
plt.show()
```
